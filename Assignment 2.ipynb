2. Load the dataset.

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['House Price India.csv']))
print(df)
```

3. Perform the Below Visualizations.
  *  Univariate Analysis




```python
summary_stats = df.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
df['Date'].value_counts().plot(kind='bar')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Bar Chart of Date')
plt.show()
```

```python
plt.hist(df['id'], bins=20)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Histogram of id')
plt.show()
```



  * Bi - Variate Analysis




```python
plt.scatter(df['number of bedrooms'], df['number of bathrooms'])
plt.xlabel('bedrooms')
plt.ylabel('bathrooms')
plt.title('Scatter Plot between bedrooms and bathrooms')
plt.show()
```

        

*   Multivariate Analysis







```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca_result = pca.fit_transform(df[['number of bedrooms', 'lot area', 'number of floors']])

explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratios:", explained_variance_ratio)
```
